```python
import serial
import time
import sys
```

```python
SERIAL_PORT = 'COM12' 
BAUD_RATE = 9600

try:
    ser = serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=1)
    time.sleep(2)
    print(f"Connected to Arduino on {SERIAL_PORT}")
except Exception as e:
    print(f"Error: Could not open serial port: {e}")
    sys.exit()
```

```python
def main():
    print("Arduino Servo Controller")
    print("Type 'q' to quit")
    
    while True:
        user_input = input("Enter degrees (0-180): ").strip()
        
        if user_input.lower() == 'q':
            print("Program ended.")
            ser.close()
            break
            
        ser.write(f"{user_input}\n".encode('utf-8'))
        time.sleep(0.1)
        while ser.in_waiting > 0:
            
            feedback = ser.readline().decode('utf-8').strip()
            print(feedback)
```

```python
if __name__ == '__main__':
    main()
```
